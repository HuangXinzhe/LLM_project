# 提高RAG性能的9种技术


## 1、提高索引数据的质量
- 索引的数据决定了RAG答案的质量
- 数据清洗的一些方法
  - 清除特殊字符、奇怪的编码、不必要的HTML标记来消除文本噪声
  - 找出与主要主题无关的文档异常值并将其删除可以通过实现一些主题提取、降维技术和数据可视化来实现这一点）
  - 使用相似性度量删除冗余文档
## 2、优化索引结构
在构建RAG时，块大小是一个关键参数，它决定了我们从矢量存储中检索的文档的长度。小的块大小可能会导致文档错过一些关键信息，而大的块大小则会引入不相关的噪声
## 3、添加元数据
将元数据与索引向量结合有助于提高搜索相关性。以下是元数据有用的一些场景：- 如果搜索项目并且以最近为标准，则可以对日期元数据进行排序；
- 如果你搜索科学论文，并且事先知道你要寻找的信息总是位于特定的部分，比如实验部分，你可以将文章部分添加为每个区块的元数据，并对其进行过滤，以仅匹配实验。
## 4、将输入查询与文档对齐
- 不优化与文档的相似性，而是优化与潜在问题的相似性。（通过文本段落可以回答的问题与输入进行相似度的比较）
## 5、混合检索
- 充分利用矢量搜素和关键字搜索等不同检索技术的优势，并将它们智能地组合在一起
## 6、重新排序
- 对检索到的文本进行重新排序，以期更有益的文本排在前面
## 7、提示压缩
- 检索到的上下文中的噪声会对RAG性能产生不利影响，在检索后应用后处理步骤来压缩不相关的上下文，突出显示重要段落，并减少整体上下文长度
## 8、HyDE
- HyDE指示LLM生成一个假设答案，利用假设答案的嵌入向量在语料库中检索回答问题的相关文本，可以有效的过滤一些不正确的细节
## 9、查询重写和扩展
- 用户输入的查询格式不一定很好，也不能完全表达与向量库中的文本有效匹配的意图。通过对输入查询的重写和扩展，可以生成更精确的查询，从而提高RAG的性能。
